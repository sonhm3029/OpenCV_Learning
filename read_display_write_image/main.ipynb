1. `imread()` để đọc ảnh
2. `imshow()` để show ảnh trong cửa sổ
3. `imwrite()` để ghi ảnh ra file

```python
import cv2 
import matplotlib.pyplot as plt

%matplotlib inline

# The function cv2.imread() is used to read an image
img_gray = cv2.imread('../assets/images/example_img.png', 0)

# The function cv2.imshow() is used to display an image in a window
cv2.imshow('gray scale image', img_gray)

# # waitKey() waits for a key press to close the window and 0 specifies indefinite loop
cv2.waitKey(0)

# The function cv2.imwrite() is use to write an image.
cv2.imwrite('gray_image.png', img_gray)
```

## 1. Reading an Image

ĐỌc image vào với hàm:

`imread(filename, flags)`

2 tham số truyền vào :

- `filename` chỉ ra đường dẫn đến file muốn đọc
- `flags` chỉ ra `how the images should be represented`, nó có rất nhiều option nhưng chủ yếu là dùng 3 cái sau:
    * `cv2.IMREAD_UNCHANGED` hoặc `-1`
    * `cv2.IMREAD_GRAYSCALE` hoặc `0`
    * `cv2.IMREAD_COLOR` hoặc `1`

Default thì `flags = 1` là đọc vào ảnh là ảnh màu với format là `BGR` mà các thư viện computer vision thì thường dùng `RGB` cho nên ta cần phải đổi lại `BGR` thành `RGB`


```python
img_BGR = cv2.imread('../assets/images/example_img.png', cv2.IMREAD_COLOR)
img_RGB = cv2.cvtColor(img_BGR, cv2.COLOR_BGR2RGB)

# matplot lib to show and compare 2 mode above
fig = plt.figure(figsize=(60, 60))
ax1 = fig.add_subplot(121)
plt.imshow(img_BGR)
ax2 = fig.add_subplot(122)
plt.imshow(img_RGB)
plt.rcParams['axes.titlesize'] = 60
ax1.set_title('BGR image')
ax2.set_title('RGB image')


plt.show()
```

## 2. Displaye Image

`cv2.imshow(window_name, image)`

- `window_name` là tên cửa sổ hiển thị image
- `image` là ảnh muốn hiển thị

## 3. Write Image

`cv2.imwrite(filename, image)`

- `filename` chỉ ra đường dẫn để lưu file
- `image` là ảnh muốn lưu ra (chú ý chế độ của ảnh khi lưu: ảnh xám, ảnh màu, ...)


**Dùng lệnh:** `cv2.destroyAllWindows` để đóng tất cả các cửa số đang mở.
