# Annotating image

Annotating image và videos có nhiều mục đích:

* Thêm info, chú thích vào ảnh/ video
* Vẽ bounding box quanh vật thể trong bài toán object detection
* Highlighting pixels với color khác nhau cho bài toán image segmentation

````python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

img = cv2.imread('../assets/images/example_img.png', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.title('Original Image')
plt.show()
````

## 1. Annotate an Image with a color line

Sử dụng hàm `line()` để vẽ:

**Syntax:**

```python
line(image, start_point, end_point, color, thickness)
```

- `image` là ảnh để vẽ
- `start_point, end_point` là 2 đầu mút của line vẽ

````python
imageLine = img.copy()

# Draw the image from point A to B
pointA = (200, 80)
pointB = (450, 80)
cv2.line(imageLine, pointA, pointB, (255, 255, 0), thickness=3)

# show image
plt.imshow(imageLine)
plt.title('Image Line')
plt.show()
````

## 2. Draw circle over an image

Sử dụng hàm `circle()` để vẽ vòng tròn

**Syntax:**

```python
circle(image, center_coordinates, radius, color, thickness)
```

Trong đó:

- `image` là ảnh để vẽ
- `center_coordinates, radius` define center của vòng tròn và radius ( độ lớn) của nó
- `color, thickness` là color và độ dày của vòng tròn.

````python
# Make copy of image
imageCircle = img.copy()

# define center of circl
circle_center = (250, 200)

# define the radius of the circle
radius = 100

# Draw a circle using circle() function
cv2.circle(imageCircle, circle_center, radius, (0, 0, 255), thickness=3, lineType=cv2.LINE_AA)

# Display the result
plt.imshow(imageCircle)
plt.title('Image circle')
plt.show()
````

## 3. Overlay an Image with a Filled Circle

Để vẽ vòng tròn filled color thì chỉ cần set `thickness=-1`

````python
# Make copy of image
imageCircle = img.copy()

# define center of circl
circle_center = (250, 200)

# define the radius of the circle
radius = 100

# Draw a circle using circle() function
cv2.circle(imageCircle, circle_center, radius, (0, 0, 255), thickness=-1, lineType=cv2.LINE_AA)

# Display the result
plt.imshow(imageCircle)
plt.title('Image circle')
plt.show()
````

## 4. Annotate images with rectangles

Sử dụng hàm `rectangle()`

**Syntax:**

```python
rectangle(image, start_point, end_point, color, thickness)
```

Trong đó:

- `start_point`: là top lep point của image
- `end_point`: là bottom right point của image

````python
# make a copy of the original image
imageRectangle = img.copy()
# define the starting and end points of the rectangle
start_point =(150,115)
end_point =(400,300)
# draw the rectangle
cv2.rectangle(imageRectangle, start_point, end_point, (0, 0, 255), thickness= 3, lineType=cv2.LINE_8) 
# display the output
plt.imshow(imageRectangle)
plt.title('Image rectangle')
plt.show()
````

## 5. Draw Ellipses on Images

Ta sử dụng hàm `ellipse()` để vẽ. Hàm này sử dụng khá giống hàm vẽ vòng trong `circle()` trừ việc thay vì có tham số `radius` thì cần các tham số khác là:

- major and minor axis length của ellipse
- rotation angle
- starting and ending angle

**Syntax:**

```python
ellipse(image, centerCoordinates, axesLength, angle, startAngle, endAngle, color, thickness)
```



````python
# make copy of the original image
imageEllipse = img.copy()

# define the center point of ellipse
ellipse_center =(415, 190)

# define major and minor axes of the ellipse
axis1 = (100, 50)
axis2 = (125, 50)

# draw ellipse
# Horizontal
cv2.ellipse(imageEllipse, ellipse_center, axis1, 0, 0, 360, (255, 0, 0), thickness=3)
# Vertical
cv2.ellipse(imageEllipse, ellipse_center, axis2, 90, 0, 360, (0, 0, 255), thickness=3)

# display the output
plt.imshow(imageEllipse)
plt.show()
````

## 6. Overlay Images with outlined and Filled up half-ellipses

````python
# make a copy of the original image
halfEllipse = img.copy()
# define the center of half ellipse
ellipse_center = (415,190)
# define the axis point
axis1 = (100,50)
# draw the Incomplete/Open ellipse, just a outline
cv2.ellipse(halfEllipse, ellipse_center, axis1, 0, 180, 360, (255, 0, 0), thickness=3)
# if you want to draw a Filled ellipse, use this line of code
cv2.ellipse(halfEllipse, ellipse_center, axis1, 0, 0, 180, (0, 0, 255), thickness=-2)
# display the output
plt.imshow(halfEllipse)
plt.show()
````

## 7. Annotate Images with text using different font faces and styled

Đê thêm text ta sử dụng hàm `putText()`

**Syntax:**

```python
putText(image, text, org, font, fontScale, color)
```

- `image` là ảnh đễ vẽ
- `text` là text muốn vẽ vào ảnh
- `org` chỉ ra start location cho top left corner của text string
- OpenCV supports several font-face styles from the Hershey font collection, and an italic font as well. Check out this list: 
  * FONT_HERSHEY_SIMPLEX        = 0,
  * FONT_HERSHEY_PLAIN          = 1,
  * FONT_HERSHEY_DUPLEX         = 2,
  * FONT_HERSHEY_COMPLEX        = 3,
  * FONT_HERSHEY_TRIPLEX        = 4,
  * FONT_HERSHEY_COMPLEX_SMALL  = 5,
  * FONT_HERSHEY_SCRIPT_SIMPLEX = 6,
  * FONT_HERSHEY_SCRIPT_COMPLEX = 7,
  * FONT_ITALIC                 = 16


````python
# make a copy of the original image
imageText  = img.copy()

# text
text = 'I am a beautiful rose'

# org: Where to put text
org = (10, 100)

# write text on input image
cv2.putText(imageText, text, org, fontFace = cv2.FONT_HERSHEY_COMPLEX, fontScale = 1.0, color = (255,225,255))

plt.imshow(imageText)
plt.show()
````
