# Color spaces in OpenCV Python



Sử dụng hàm `cvtColor()` để convert giữa các hệ màu khác nhau trong OpenCV.

```python
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
bright = cv2.imread('../assets/images/example_img.png')
```

## 1. LAB Color-Space

- `L`- Lightness (intensity)
- `a`- color component ranging from Greean to Magenta
- `b`- color component ranging from Blue to Yellow

Hệ màu `LAB` khác so với hệ `RGB` đó là `RGB` có 3 kênh màu, trong đó mỗi kênh đều là mã hóa brightness. Còn đối với hệ màu `LAB` thì channel `L` is independent of color information and encodes brightness only. The other two channels encode color.

Nó có các properties sau:

* Perceptually uniform color space which approximates how we perceive color. (Không gian màu đồng nhất về mặt tri giác, gần đúng với cách chúng ta cảm nhận màu sắc.)
* Independent of device( capturing or displaying). Độc lập với thiết bị (chụp hoặc hiển thị).
* Used extensively in Adobe Photoshop
* Is related to the RGB color space by a complex transformation equation

```python
brightLAB = cv2.cvtColor(bright, cv2.COLOR_BGR2LAB)

# show image
plt.subplot(1,2,1)
plt.imshow(bright)
plt.title("Original Image")
plt.subplot(1,2,2)
plt.imshow(brightLAB)
plt.title("LAB image")
plt.show()
```

## 2. The YCrCb Color-Space
1. Y- Luminance or Luma component
2. Cr= R - Y (how far is the red component from Luma)
3. Cb = B - Y (how far is the blue component from Luma).

This color space has the following properties

- Separates the lumninance and chrominance components into different channels
- Mostly used in compression (of Cr and Cb components) for TV Transmission
- Device dependent.



```python
brightYCB = cv2.cvtColor(bright, cv2.COLOR_BGR2YCrCb)

# show image
plt.subplot(1,2,1)
plt.imshow(bright)
plt.title("Original Image")
plt.subplot(1,2,2)
plt.imshow(brightYCB)
plt.title("YCrCb Image")
plt.show()
```

## 3. The HSV Color Space

- `H`- Hue (Dominant Wavelength)
- `S`- Saturation (Purity / shades of the color)
- `V`- Value (Intensity)

Let's enmerate some of its properties

* Hệ màu này sử dụng channel color (H)
* Device dependent



```python
brightHSV = cv2.cvtColor(bright, cv2.COLOR_BGR2HSV)


# show image
plt.subplot(1,2,1)
plt.imshow(bright)
plt.title("Original Image")
plt.subplot(1,2,2)
plt.imshow(brightHSV)
plt.title("HSV Image")
plt.show()
```
