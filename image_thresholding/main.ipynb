# Thresholding Image


````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

src = cv2.imread('../assets/images/example_img.png',cv2.IMREAD_GRAYSCALE)

# cv2.imshow("ok", src)
# cv2.waitKey(0)

plt.imshow(src,cmap=plt.cm.gray)
plt.title("Original grayscale image")
````

## 1. Global thresholding

`Global threshold` là khi threshold rule được apply như nhau cho tất cả các pixel của image và threshold value là 1 giá trị fixed.

`Global threshold` nhận vào ảnh đầu vào (`src`) và threshold value (`thresh`), đầu ra ảnh (`dst`) là kết quả của việc so sánh pixel intensity `src(x, y)` của ảnh đầu vào với threshold. Nếu `src(x, y)` > `thresh` thì `dst(x, y)` được gán bởi value nào đó. Nếu không thì `dst(x, y)` được gán bằng một giá trị khác.

`Simplest form` của global thresholding được gọi là `Binary Thresholding`.

Mô tả thuật toán như sau:

```python
maxValue
if src(x, y) > thresh
    dst(x, y) =maxValue
else:
    dst(x, y) = 0
```

## 2. Binary thresholding

````python
thresh = 0
maxValue = 255

# Basic apply binary thresholding
th, dst = cv2.threshold(src, thresh, maxValue, cv2.THRESH_BINARY)

print(f"Thresh hold: {th}")

plt.subplot(1,2,1)
plt.imshow(src, cmap=plt.cm.gray)
plt.title("Original image")

plt.subplot(1,2,2)
plt.imshow(dst, cmap=plt.cm.gray)
plt.title("binary threshold image")
````

````python
thresh = 100
maxValue = 255

# Basic apply binary thresholding
th, dst = cv2.threshold(src, thresh, maxValue, cv2.THRESH_BINARY)

print(f"Thresh hold: {th}")

plt.subplot(1,2,1)
plt.imshow(src, cmap=plt.cm.gray)
plt.title("Original image")

plt.subplot(1,2,2)
plt.imshow(dst, cmap=plt.cm.gray)
plt.title("binary threshold image")
````

````python
thresh = 150
maxValue = 200

# Basic apply binary thresholding
th, dst = cv2.threshold(src, thresh, maxValue, cv2.THRESH_BINARY)

print(f"Thresh hold: {th}")

plt.subplot(1,2,1)
plt.imshow(src, cmap=plt.cm.gray)
plt.title("Original image")

plt.subplot(1,2,2)
plt.imshow(dst, cmap=plt.cm.gray)
plt.title("binary threshold image")
````

## 3. Inverse-Binary thresholding

Đây là dạng ngược lại của Binary thresholding

```python
# Inverse Binary Threshold
if src(x,y) > thresh
  dst(x,y) = 0
else
  dst(x,y) = maxValue
```

````python
thresh = 150
maxValue = 200

# Basic apply binary thresholding
th1, dst1 = cv2.threshold(src, thresh, maxValue, cv2.THRESH_BINARY_INV)
th2, dst2 = cv2.threshold(src, thresh, maxValue, cv2.THRESH_BINARY)

print(f"Thresh hold: {th1}")

plt.rcParams['axes.titlesize'] = 40
plt.rcParams['figure.figsize'] = (40, 40)


plt.subplot(1,3,1)
plt.imshow(src, cmap=plt.cm.gray)
plt.title("Original image")

plt.subplot(1,3,2)
plt.imshow(dst1, cmap=plt.cm.gray)
plt.title("inverse binary threshold image")

plt.subplot(1,3,3)
plt.imshow(dst2, cmap=plt.cm.gray)
plt.title("binary threshold image")
````

## 4. Truncate Thresholding

Với dạng này thì:

- destination pixel được gán bằng `thresh` nếu source pixel có giá trị lớn hơn `thresh`
- Nếu không thì nó được gán bằng chính source pixel value (chính nó)
- `maxValue` không dùng trong dạng này

```python
# Truncate Threshold
if src(x,y) > thresh
  dst(x,y) = thresh
else
  dst(x,y) = src(x,y)
```

````python
thresh = 150
maxValue = 200

# Basic apply binary thresholding
th, dst = cv2.threshold(src, thresh, maxValue, cv2.THRESH_TRUNC)

print(f"Thresh hold: {th}")

plt.subplot(1,2,1)
plt.imshow(src, cmap=plt.cm.gray)
plt.title("Original image")

plt.subplot(1,2,2)
plt.imshow(dst, cmap=plt.cm.gray)
plt.title("Truncate threshold image")
````

## 5. Threshold to Zero

Trong dạng này:

- Destination pixel value được set bằng chính nó nếu nó lớn hơn `thresh`
- Nếu không thì được set = 0
- Không dùng `maxValue` trong dạng này

```python
# Threshold to Zero
if src(x,y) > thresh
  dst(x,y) = src(x,y)
else
  dst(x,y) = 0
```

````python
thresh = 150
maxValue = 200

# Basic apply binary thresholding
th, dst = cv2.threshold(src, thresh, maxValue, cv2.THRESH_TOZERO)

print(f"Thresh hold: {th}")

plt.subplot(1,2,1)
plt.imshow(src, cmap=plt.cm.gray)
plt.title("Original image")

plt.subplot(1,2,2)
plt.imshow(dst, cmap=plt.cm.gray)
plt.title("To zero threshold image")
````

## 6. Inverted Threshold to Zero

Dạng ngược lại của to Zero threshold

````python
thresh = 150
maxValue = 200

# Basic apply binary thresholding
th, dst = cv2.threshold(src, thresh, maxValue, cv2.THRESH_TOZERO_INV)

print(f"Thresh hold: {th}")

plt.subplot(1,2,1)
plt.imshow(src, cmap=plt.cm.gray)
plt.title("Original image")

plt.subplot(1,2,2)
plt.imshow(dst, cmap=plt.cm.gray)
plt.title("Inverse to zero threshold image")
````
