# Thresholding Image


````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

src = cv2.imread('../assets/images/example_img.png',cv2.IMREAD_GRAYSCALE)

# cv2.imshow("ok", src)
# cv2.waitKey(0)

plt.imshow(src,cmap=plt.cm.gray)
plt.title("Original grayscale image")
````

## 1. Global thresholding

`Global threshold` l khi threshold rule c apply nh nhau cho tt c cc pixel ca image v threshold value l 1 gi tr fixed.

`Global threshold` nhn vo nh u vo (`src`) v threshold value (`thresh`), u ra nh (`dst`) l kt qu ca vic so snh pixel intensity `src(x, y)` ca nh u vo vi threshold. Nu `src(x, y)` > `thresh` th `dst(x, y)` c gn bi value no . Nu khng th `dst(x, y)` c gn bng mt gi tr khc.

`Simplest form` ca global thresholding c gi l `Binary Thresholding`.

M t thut ton nh sau:

```python
maxValue
if src(x, y) > thresh
    dst(x, y) =maxValue
else:
    dst(x, y) = 0
```

## 2. Binary thresholding

````python
thresh = 0
maxValue = 255

# Basic apply binary thresholding
th, dst = cv2.threshold(src, thresh, maxValue, cv2.THRESH_BINARY)

print(f"Thresh hold: {th}")

plt.subplot(1,2,1)
plt.imshow(src, cmap=plt.cm.gray)
plt.title("Original image")

plt.subplot(1,2,2)
plt.imshow(dst, cmap=plt.cm.gray)
plt.title("binary threshold image")
````

````python
thresh = 100
maxValue = 255

# Basic apply binary thresholding
th, dst = cv2.threshold(src, thresh, maxValue, cv2.THRESH_BINARY)

print(f"Thresh hold: {th}")

plt.subplot(1,2,1)
plt.imshow(src, cmap=plt.cm.gray)
plt.title("Original image")

plt.subplot(1,2,2)
plt.imshow(dst, cmap=plt.cm.gray)
plt.title("binary threshold image")
````

````python
thresh = 150
maxValue = 200

# Basic apply binary thresholding
th, dst = cv2.threshold(src, thresh, maxValue, cv2.THRESH_BINARY)

print(f"Thresh hold: {th}")

plt.subplot(1,2,1)
plt.imshow(src, cmap=plt.cm.gray)
plt.title("Original image")

plt.subplot(1,2,2)
plt.imshow(dst, cmap=plt.cm.gray)
plt.title("binary threshold image")
````

## 3. Inverse-Binary thresholding

y l dng ngc li ca Binary thresholding

```python
# Inverse Binary Threshold
if src(x,y) > thresh
  dst(x,y) = 0
else
  dst(x,y) = maxValue
```

````python
thresh = 150
maxValue = 200

# Basic apply binary thresholding
th1, dst1 = cv2.threshold(src, thresh, maxValue, cv2.THRESH_BINARY_INV)
th2, dst2 = cv2.threshold(src, thresh, maxValue, cv2.THRESH_BINARY)

print(f"Thresh hold: {th1}")

plt.rcParams['axes.titlesize'] = 40
plt.rcParams['figure.figsize'] = (40, 40)


plt.subplot(1,3,1)
plt.imshow(src, cmap=plt.cm.gray)
plt.title("Original image")

plt.subplot(1,3,2)
plt.imshow(dst1, cmap=plt.cm.gray)
plt.title("inverse binary threshold image")

plt.subplot(1,3,3)
plt.imshow(dst2, cmap=plt.cm.gray)
plt.title("binary threshold image")
````

## 4. Truncate Thresholding

Vi dng ny th:

- destination pixel c gn bng `thresh` nu source pixel c gi tr ln hn `thresh`
- Nu khng th n c gn bng chnh source pixel value (chnh n)
- `maxValue` khng dng trong dng ny

```python
# Truncate Threshold
if src(x,y) > thresh
  dst(x,y) = thresh
else
  dst(x,y) = src(x,y)
```

````python
thresh = 150
maxValue = 200

# Basic apply binary thresholding
th, dst = cv2.threshold(src, thresh, maxValue, cv2.THRESH_TRUNC)

print(f"Thresh hold: {th}")

plt.subplot(1,2,1)
plt.imshow(src, cmap=plt.cm.gray)
plt.title("Original image")

plt.subplot(1,2,2)
plt.imshow(dst, cmap=plt.cm.gray)
plt.title("Truncate threshold image")
````

## 5. Threshold to Zero

Trong dng ny:

- Destination pixel value c set bng chnh n nu n ln hn `thresh`
- Nu khng th c set = 0
- Khng dng `maxValue` trong dng ny

```python
# Threshold to Zero
if src(x,y) > thresh
  dst(x,y) = src(x,y)
else
  dst(x,y) = 0
```

````python
thresh = 150
maxValue = 200

# Basic apply binary thresholding
th, dst = cv2.threshold(src, thresh, maxValue, cv2.THRESH_TOZERO)

print(f"Thresh hold: {th}")

plt.subplot(1,2,1)
plt.imshow(src, cmap=plt.cm.gray)
plt.title("Original image")

plt.subplot(1,2,2)
plt.imshow(dst, cmap=plt.cm.gray)
plt.title("To zero threshold image")
````

## 6. Inverted Threshold to Zero

Dng ngc li ca to Zero threshold

````python
thresh = 150
maxValue = 200

# Basic apply binary thresholding
th, dst = cv2.threshold(src, thresh, maxValue, cv2.THRESH_TOZERO_INV)

print(f"Thresh hold: {th}")

plt.subplot(1,2,1)
plt.imshow(src, cmap=plt.cm.gray)
plt.title("Original image")

plt.subplot(1,2,2)
plt.imshow(dst, cmap=plt.cm.gray)
plt.title("Inverse to zero threshold image")
````
